```python
import method_fibonacci
import powell_quad
import hook_jeevs
import nelder_m
import rosenbrok
import powell_sopr_n
```

```python
cache = {}
def func_1(x):
    if x not in cache:
        cache[x] = x**2
    return cache[x]

def func_2(x):
    # -0.472
    if x not in cache:
        cache[x] = (6 * x + 3)**2 - 2*x - 1
    return cache[x]

def func_3(x):
    #3
    if x not in cache:
        cache[x] = x**2 - (6*x) + 14
    return cache[x]
funcs = [func_1, func_2, func_3]
intervals = [(-6, 6), (-6, 6), (-60, 60)]
reals=[0, func_2(-0.472), func_3(3)]
```

```python
methods = [method_fibonacci.fibonacci_method, powell_quad.quad_inter_method]
```

```python
df1 = pd.DataFrame(columns=['epsilon', 'fibonacci_method отклонение', 'powell_quad_inter отклонение', 'fibonacci_method количество вычислений', 'powell_quad_inter количество вычислений'])
df1['epsilon'] = [0.01, 0.0001, 0.000001]
df2 = df1.copy()
df3 = df1.copy()
dfs = [df1, df2, df3]
```

```python
for i in range(len(dfs)):
    func = funcs[i]
    for j in range(len(df1['epsilon'])):
        min = method_fibonacci.fibonacci_method(func, intervals[i][0], intervals[i][1], eps=df1['epsilon'][j])
        dfs[i]['fibonacci_method отклонение'][j] = abs(reals[i] - func(min))
        dfs[i]['fibonacci_method количество вычислений'][j] = len(cache)
        cache = {}

        min = powell_quad.quad_inter_method(func, intervals[i][0], eps1=df1['epsilon'][j], eps2=df1['epsilon'][j])
        dfs[i]['powell_quad_inter отклонение'][j] = abs(reals[i] - func(min))
        dfs[i]['powell_quad_inter количество вычислений'][j] = len(cache)
        cache = {}
```
