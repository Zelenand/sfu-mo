```python
import pandas as pd
import numpy as np
from numpy.linalg import inv
```

```python
from markwardt import markwardt
from newton import newton
from newton_rafson import newton_rafson
```

```python
cache_0 = {}
cache_1 = {}
cache_2 = {}


def func_1(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = x[0] ** 2 + x[1] ** 2
    return cache_0[x]

def func_1_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 2 * x[0]
    return cache_1[x]

def func_1_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 2 * x[1]
    return cache_2[x]

def hesse_1(x):
    h = np.array([[2, 0], [0, 2]])
    h_1 = inv(h)
    return h, h_1

def func_2(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = (x[0] - 1)**2 + (x[1] + x[0])**2
    return cache_0[x]

def func_2_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 4 * x[0] + 2 * x[1] - 2
    return cache_1[x]

def func_2_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 2 * x[0] + 2 * x[1]
    return cache_2[x]

def hesse_2(x):
    h = np.array([[4, 2], [2, 2]])
    h_1 = inv(h)
    return h, h_1

def func_3(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = x[0]**2 + 4 * x[0] * x[1] + 18 * (x[1]**2)
    return cache_0[x]

def func_3_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 2 * x[0] + 4 * x[1]
    return cache_1[x]

def func_3_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 4 * x[0] + 36 * x[1]
    return cache_2[x]

def hesse_3(x):
    h = np.array([[2, 4], [4, 36]])
    h_1 = inv(h)
    return h, h_1

funcs = [(func_1, func_1_d1, func_1_d2, hesse_1), (func_2, func_2_d1, func_2_d2, hesse_2), (func_3, func_3_d1, func_3_d2, hesse_3)]
reals = [0, func_2((1, -1)), func_3((0, 0))]

df1 = pd.DataFrame(columns=['epsilon', 'newton отклонение', 'newton_rafson отклонение', 'markwardt отклонение', 'newton количество вычислений', 'newton_rafson количество вычислений', 'markwardt количество вычислений'])
df1['epsilon'] = [0.1, 0.0001, 0.0000001]
M=300
df2 = df1.copy()
df3 = df1.copy()
dfs = [df1, df2, df3]

for i in range(len(dfs)):
    funcs_ = funcs[i]
    for j in range(len(df1['epsilon'])):
        print(i, j)
        print(1)
        x0 = (2, 2)
        if i == 1: x0 = (10, 10)
        elif i == 2: x0 = (100, 100)
        min = newton(funcs_[0], funcs_[1], funcs_[2], funcs_[3], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['newton отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['newton количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = newton_rafson(funcs_[0], funcs_[1], funcs_[2], funcs_[3], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['newton_rafson отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['newton_rafson количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = markwardt(funcs_[0], funcs_[1], funcs_[2], funcs_[3], x0, eps1=df1['epsilon'][j], M=M)
        dfs[i]['markwardt отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['markwardt количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}
```
