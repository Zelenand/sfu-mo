```python
import pandas as pd
```

```python
from gradient_descent_constant_stepsize import gradient_descent_constant_stepsize
from gradient_descent import gradient_descent
from coordinat_descent import coordinat_descent
from gauss_zeidel import gauss_zeidel
from fletcher_rivs import fletcher_rivs
from davidon_fletcher_powell import davidon_fletcher_powell
```

```python
cache_0 = {}
cache_1 = {}
cache_2 = {}


def func_1(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = x[0] ** 2 + x[1] ** 2
    return cache_0[x]

def func_1_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 2 * x[0]
    return cache_1[x]

def func_1_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 2 * x[1]
    return cache_2[x]


def func_2(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = (x[0] - 1)**2 + (x[1] + x[0])**2
    return cache_0[x]

def func_2_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 4 * x[0] + 2 * x[1] - 2
    return cache_1[x]

def func_2_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 2 * x[0] + 2 * x[1]
    return cache_2[x]


def func_3(x):
    x = tuple(x)
    if x not in cache_0:
        cache_0[x] = x[0]**2 + 4 * x[0] * x[1] + 18 * (x[1]**2)
    return cache_0[x]

def func_3_d1(x):
    x = tuple(x)
    if x not in cache_1:
        cache_1[x] = 2 * x[0] + 4 * x[1]
    return cache_1[x]

def func_3_d2(x):
    x = tuple(x)
    if x not in cache_2:
        cache_2[x] = 4 * x[0] + 36 * x[1]
    return cache_2[x]

funcs = [(func_1, func_1_d1, func_1_d2), (func_2, func_2_d1, func_2_d2), (func_3, func_3_d1, func_3_d2)]
reals = [0, func_2((1, -1)), func_3((0, 0))]

df1 = pd.DataFrame(columns=['epsilon', 'gradient_descent_constant_stepsize отклонение', 'gradient_descent отклонение', 'coordinat_descent отклонение', 'gauss_zeidel отклонение', 'fletcher_rivs отклонение', 'davidon_fletcher_powell отклонение', 'gradient_descent_constant_stepsize количество вычислений', 'gradient_descent количество вычислений', 'coordinat_descent количество вычислений', 'gauss_zeidel количество вычислений', 'fletcher_rivs количество вычислений', 'davidon_fletcher_powell количество вычислений'])
df1['epsilon'] = [0.1, 0.0001, 0.0000001]
M=300
df2 = df1.copy()
df3 = df1.copy()
dfs = [df1, df2, df3]

for i in range(len(dfs)):
    funcs_ = funcs[i]
    for j in range(len(df1['epsilon'])):
        print(i, j)
        print(1)
        x0 = (2, 2)
        if i == 1: x0 = (10, 10)
        elif i == 2: x0 = (100, 100)
        min = gradient_descent_constant_stepsize(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['gradient_descent_constant_stepsize отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['gradient_descent_constant_stepsize количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = gradient_descent(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['gradient_descent отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['gradient_descent количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = coordinat_descent(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['coordinat_descent отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['coordinat_descent количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = gauss_zeidel(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['gauss_zeidel отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['gauss_zeidel количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = fletcher_rivs(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['fletcher_rivs отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['fletcher_rivs количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

        min = davidon_fletcher_powell(funcs_[0], funcs_[1], funcs_[2], x0, eps1=df1['epsilon'][j], eps2=df1['epsilon'][j], M=M)
        dfs[i]['davidon_fletcher_powell отклонение'][j] = abs(reals[i] - funcs_[0](min))
        dfs[i]['davidon_fletcher_powell количество вычислений'][j] = len(cache_0) + len(cache_1) + len(cache_2)
        cache_0 = {}
        cache_1 = {}
        cache_2 = {}

```
